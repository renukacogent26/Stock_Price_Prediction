```python
X=df.values
X=df.drop(columns=['Adj Close', 'Close'], axis=1)
X.head()
```

```python
y=df['Close']
print(X.shape)
print(y.shape)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X_train,X_test,y_train,y_test=train_test_split(X,y, random_state=42)


rt_clf=RandomForestRegressor(max_depth=2, random_state=42)
rt_clf.fit(X_train,y_train)

y_predict=rt_clf.predict(X_test)
print(y_predict)
print(y_predict.shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import numpy as np

mae=mean_absolute_error(y_test, y_predict)
mse=mean_squared_error(y_test,y_predict)
root_mse=np.sqrt(mse)
```

```python
print(f"The mean absolute error is","%.2f" % mae)
print(f"The mean squared error is","%.2f" % mse)
print(f"The R_squared is", "%.2f" % root_mse)
```

```python
prediction=pd.DataFrame(y_predict)
prediction['Date']=df.index[-253:]
prediction.columns=["Predicted", "Date"]
columns_titles = ["Date","Predicted"]
prediction=prediction.reindex(columns=columns_titles)
prediction
```

```python
from matplotlib import pyplot

one_year_prediction=prediction
plt.figure(figsize=(15,8))
one_year_prediction['Predicted'].plot()
pyplot.title("One year prediction", fontsize=14)
pyplot.xlabel("Days", fontsize=14)
pyplot.ylabel("Price", fontsize=14)
plt.legend(loc="upper right")
pyplot.show
```
